## Step 1: Getting The Necessary Data

### Import the necessary Python libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Set the value of columns and rows displayed

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 10)
```

### Set the URL to the GitHub repository and load the data

```python
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv"
covid_raw = pd.read_csv(url)
```

## Step 2: Filtering The Data

### Subselect New York

```python
is_state = covid_raw["Province_State"] == "New York"
data = covid_raw[is_state]
```

### Remove the unnecessary columns

```python
data = data.drop(['UID', 'iso2', 'iso3', 'code3', 'FIPS', 'Province_State', 'Country_Region', 'Lat', 'Long_', 'Combined_Key'], axis=1)
```

### Set the index to the 'Admin2' column and Rename to 'Counties'

```python
data.set_index("Admin2", inplace = True)
data.index.names = ['Counties']
```

### Display the first 10 rows of the filtered dataframe

```python
data.head(n=10)
```

## Step 3: Plotting the Data

### Plot all of the data for the specified state

```python
data.T.loc['1/22/20':,:].plot(figsize=(40,20))
plt.show()
```

### Plot just March data for a specific state

```python
data.T.loc['3/1/20':,:].plot(figsize=(40,20))
plt.show()
```

### Plot March data for the specified counties in the specified state

```python
data.T.loc['3/1/20':,["Bronx", "Kings", "Nassau", "New York", "Queens", "Richmond", "Suffolk", "Westchester"]].plot(figsize=(16, 8))
plt.show()
```

## Step 4: Add Totals Column to Data Frame

### Display the first 5 rows of the filtered dataframe again

```python
data.head()
```

### Sum the values for each county (i.e. by row)

```python
data_totals = data.sum(axis=1)
```

### Add a column called Totals with the county totals

```python
data['Totals'] = data_totals
```

### Display the first 5 rows of the updated dataframe

```python
data.head()
```

## Step 5: Plot The Top Eight Counties of Covid-19 Cases

### Filter the top 8 county totals

```python
top = data.nlargest(8, 'Totals')
```

### Display the first 5 rows of the new dataframe

```python
top.head(n=8)
```

### Display a horizontal bar chart with title, labels, and values for each bar


```python
barplot = top.loc[:, 'Totals'].plot(kind = 'barh', fontsize = 20, figsize = (24, 8))
for i, v in enumerate(top.Totals):
    barplot.text(v + 3, i, " "+str(v), va = 'center', color = 'black', fontweight = 'bold')
plt.title('Total Covid-19 cases for top eight New York counties', fontsize = 24)
plt.xlabel('Total cases', fontsize = 24)
plt.ylabel('County', fontsize = 24)

# df.plot(kind="barh", x="state", y=["positive", "death"], fontsize=100, figsize=(150, 100))
# plt.legend(prop={"size":150})
```
